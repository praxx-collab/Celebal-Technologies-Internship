# Question 1: Orders with Sales Greater than Average

## Objective

Retrieve all orders where the sales amount is greater than the average sales using a subquery.

```python
query = """
SELECT
    `Order ID`,
    `Customer ID`,
    Sales
FROM orders
WHERE Sales >
(
    SELECT AVG(Sales)
    FROM orders
);
"""

run_query(query)
```

### Observation

This query retrieves all orders whose sales value is greater than the average sales across all orders. It demonstrates the use of a subquery in the WHERE clause for comparison.

# Question 2: Highest Sales Order for Each Customer

## Objective

Identify the highest sales order placed by each customer using a correlated subquery.

```python
query = """
SELECT
    o.`Order ID`,
    o.`Customer ID`,
    o.Sales
FROM orders o
JOIN
(
    SELECT
        `Customer ID`,
        MAX(Sales) AS Max_Sales
    FROM orders
    GROUP BY `Customer ID`
) m
ON o.`Customer ID` = m.`Customer ID`
AND o.Sales = m.Max_Sales;
"""

run_query(query)
```

### Observation

The query identifies the highest-value order placed by each customer using a correlated subquery.

# Question 3: Calculate Total Sales per Customer using a CTE

## Objective

Use a Common Table Expression (CTE) to calculate the total sales for each customer.

```python
query = """
WITH CustomerSales AS
(
    SELECT
        `Customer ID`,
        ROUND(SUM(Sales), 2) AS Total_Sales
    FROM orders
    GROUP BY `Customer ID`
)

SELECT *
FROM CustomerSales;
"""

run_query(query)
```

### Observation

The Common Table Expression (CTE) calculates the total sales for each customer, making the query more readable and reusable for further analysis.

# Question 4: Customers with Above Average Total Sales

## Objective

Identify customers whose total sales are greater than the average customer sales using a CTE and subquery.

```python
query = """
WITH CustomerSales AS
(
    SELECT
        `Customer ID`,
        ROUND(SUM(Sales), 2) AS Total_Sales
    FROM orders
    GROUP BY `Customer ID`
)

SELECT *
FROM CustomerSales
WHERE Total_Sales >
(
    SELECT AVG(Total_Sales)
    FROM CustomerSales
);
"""

run_query(query)
```

### Observation

The query identifies customers whose total sales exceed the average customer sales, helping to recognize high-value customers.

# Question 5: Rank Customers based on Total Sales

## Objective

Use the RANK() window function to rank customers according to their total sales.

```python
query = """
WITH CustomerSales AS
(
    SELECT `Customer ID`, ROUND(SUM(Sales), 2) AS Total_Sales
    FROM orders
    GROUP BY `Customer ID`
)

SELECT
    `Customer ID`,
    Total_Sales,
    RANK() OVER (ORDER BY Total_Sales DESC) AS Customer_Rank
FROM CustomerSales;
"""

run_query(query)
```

### Observation

The RANK() window function assigns rankings based on total sales, allowing easy identification of top-performing customers.

# Question 6: Assign Row Numbers to Orders

## Objective

Use the ROW_NUMBER() window function to assign a unique sequential number to each order.

```python
query = """
SELECT `Customer ID`,`Order ID`,`Order Date`, Sales,
    ROW_NUMBER() OVER
    (
        PARTITION BY `Customer ID`
        ORDER BY `Order Date`
    ) AS Order_Number
FROM orders;
"""

run_query(query)
```

### Observation

ROW_NUMBER() assigns a unique sequential number to each row, which is useful for ordering and identifying records.

# Question 7: Top 3 Customers by Sales

## Objective

Retrieve the top three customers based on total sales using window functions.

```python
query = """
WITH CustomerSales AS
(
    SELECT `Customer ID`, ROUND(SUM(Sales), 2) AS Total_Sales
    FROM orders
    GROUP BY `Customer ID`
),

CustomerRanking AS
(
    SELECT `Customer ID`, Total_Sales,
        RANK() OVER (ORDER BY Total_Sales DESC) AS Customer_Rank
    FROM CustomerSales
)

SELECT *
FROM CustomerRanking
WHERE Customer_Rank <= 3;
"""

run_query(query)
```

### Observation

The query identifies the top three customers with the highest sales, demonstrating the practical use of ranking functions for business analysis.

# Final Combined Query

## Objective

Combine JOIN, Common Table Expressions (CTEs), and Window Functions to generate a ranked customer sales report showing customer names, total sales, and customer rankings.

```python
query = """
WITH CustomerSales AS
(
    SELECT `Customer ID`, ROUND(SUM(Sales), 2) AS Total_Sales
    FROM orders
    GROUP BY `Customer ID`
)

SELECT c.`Customer Name`, cs.Total_Sales,
    RANK() OVER (ORDER BY cs.Total_Sales DESC) AS Customer_Rank
FROM CustomerSales cs
JOIN customers c
ON cs.`Customer ID` = c.`Customer ID`
ORDER BY Customer_Rank;
"""

run_query(query)
```

### Observation

This query combines JOINs, CTEs, and Window Functions to produce a comprehensive customer sales report. It ranks customers based on their total sales and demonstrates the practical use of multiple advanced SQL concepts in a single query.

## Business Analysis using the Superstore Dataset

### Objective

Apply advanced SQL concepts to solve real-world business problems using the Superstore dataset.

### Business Questions

1. Identify the Top 5 Customers based on Total Sales.
2. Identify the Bottom 5 Customers based on Total Sales.
3. Find Customers Who Placed Only One Order.
4. Identify Customers with Above Average Total Sales.
5. Find the Highest Order Value for Each Customer.

```python
from IPython.display import display

print("="*80)
print("Business Question 1: Top 5 Customers based on Total Sales")
print("="*80)

query = """
SELECT
    `Customer ID`,
    `Customer Name`,
    ROUND(SUM(Sales),2) AS Total_Sales
FROM superstore
GROUP BY `Customer ID`, `Customer Name`
ORDER BY Total_Sales DESC
LIMIT 5;
"""

display(run_query(query))


print("="*80)
print("Business Question 2: Bottom 5 Customers based on Total Sales")
print("="*80)

query = """
SELECT
    `Customer ID`,
    `Customer Name`,
    ROUND(SUM(Sales),2) AS Total_Sales
FROM superstore
GROUP BY `Customer ID`, `Customer Name`
ORDER BY Total_Sales ASC
LIMIT 5;
"""

display(run_query(query))


print("="*80)
print("Business Question 3: Customers Who Placed Only One Order")
print("="*80)

query = """
SELECT
    `Customer ID`,
    `Customer Name`,
    COUNT(DISTINCT `Order ID`) AS Total_Orders
FROM superstore
GROUP BY `Customer ID`, `Customer Name`
HAVING COUNT(DISTINCT `Order ID`) = 1;
"""

display(run_query(query))


print("="*80)
print("Business Question 4: Customers with Above Average Total Sales")
print("="*80)

query = """
WITH CustomerSales AS
(
    SELECT
        `Customer ID`,
        `Customer Name`,
        ROUND(SUM(Sales),2) AS Total_Sales
    FROM superstore
    GROUP BY `Customer ID`, `Customer Name`
)

SELECT *
FROM CustomerSales
WHERE Total_Sales >
(
    SELECT AVG(Total_Sales)
    FROM CustomerSales
)
ORDER BY Total_Sales DESC;
"""

display(run_query(query))


print("="*80)
print("Business Question 5: Highest Order Value for Each Customer")
print("="*80)

query = """
SELECT
    `Customer ID`,
    `Customer Name`,
    ROUND(MAX(Sales),2) AS Highest_Order_Value
FROM superstore
GROUP BY `Customer ID`, `Customer Name`
ORDER BY Highest_Order_Value DESC;
"""

display(run_query(query))
```
