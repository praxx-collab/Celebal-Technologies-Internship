```python
import pandas as pd

df = pd.read_csv(r"C:\Users\prapt\OneDrive\Desktop\archive (1)\Combined_dataset.csv")

print(df)
```

```python
print("\n Columns:")
print(df.columns)
```

```python
print("\n Dataset Information :")
print(df.info())
```

```python
print("Missing Values:")
print(df.isnull().sum())
```

```python
df["discount"] = df["discount"].fillna(0)
print(df.isnull().sum())
```

```python
df["seller_name"] = df["seller_name"].fillna("Unknown")
df["seller_information"] = df["seller_information"].fillna("Unknown")
df["videos"] = df["videos"].fillna("Not Available")
df["variations"] = df["variations"].fillna("No Variations")
df["what_customers_said"] = df["what_customers_said"].fillna("No Reviews")

df.isnull().sum()
```

```python
filtered_df = df[df["rating"] > 4.0]

filtered_df.head()
```

```python
filtered_df.shape
```

```python
selected_df = filtered_df[
    [
        "product_id",
        "title",
        "rating",
        "initial_price",
        "discount",
        "final_price",
        "seller_name",
        "category"
    ]
]

selected_df.head()
```

```python
selected_df.shape
```

```python
selected_df.duplicated().sum()
```

```python
cleaned_df = selected_df.drop_duplicates()
```

```python
cleaned_df.shape
```

```python
cleaned_df["final_price"].dtype
```

```python
cleaned_df["final_price"].head(10)
```

```python
cleaned_df["final_price"] = (
    cleaned_df["final_price"]
    .str.replace('"', '', regex=False)
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(float)
)
```

```python
cleaned_df["final_price"].head()
```

```python
cleaned_df["final_price"].dtype
```

```python
cleaned_df["discount_amount"] = (
    cleaned_df["initial_price"] - cleaned_df["final_price"]
)
```

```python
cleaned_df[["initial_price","final_price","discount_amount"]].head()
```

```python
cleaned_df["total_amount"] = cleaned_df["final_price"]
```

```python
cleaned_df[["final_price", "total_amount"]].head()
```

```python
cleaned_df.to_csv(
    "ecommerce_dataset_cleaned.csv",
    index=False
)

print("CSV file saved successfully!")
```
