```python
query = """
SELECT COUNT(*) AS Total_Records
FROM superstore;
"""

pd.read_sql(query, connection)
```

```python
def run_query(query):
    return pd.read_sql(query, connection)
```

```python
run_query("""
SELECT
    ROUND(SUM(Sales),2) AS Total_Sales,
    ROUND(SUM(Profit),2) AS Total_Profit,
    SUM(Quantity) AS Total_Quantity
FROM superstore;
""")
```

### Observation

The dataset contains 9,694 records. The above query summarizes the overall sales, profit, and quantity sold, providing a quick overview of business performance.

```python
run_query("""
SELECT
    COUNT(DISTINCT Category) AS Categories,
    COUNT(DISTINCT Region) AS Regions,
    COUNT(DISTINCT Segment) AS Segments
FROM superstore;
""")
```

### Observation

The dataset contains 3 product categories, 4 sales regions, and 3 customer segments.

```python
run_query("""
SELECT
    Region,
    ROUND(SUM(Sales),2) AS Total_Sales
FROM superstore
GROUP BY Region
ORDER BY Total_Sales DESC;
""")
```

### Observation

The West region recorded the highest total sales among all regions.

```python
print("Sales by Category")
display(run_query("""
SELECT
    Category,
    ROUND(SUM(Sales),2) AS Total_Sales
FROM superstore
GROUP BY Category
ORDER BY Total_Sales DESC;
"""))

print("\nTop 10 Products by Sales")
display(run_query("""
SELECT
    `Product Name`,
    ROUND(SUM(Sales),2) AS Total_Sales
FROM superstore
GROUP BY `Product Name`
ORDER BY Total_Sales DESC
LIMIT 10;
"""))

print("\nTop 5 Customers by Sales")
display(run_query("""
SELECT
    `Customer Name`,
    ROUND(SUM(Sales),2) AS Total_Sales
FROM superstore
GROUP BY `Customer Name`
ORDER BY Total_Sales DESC
LIMIT 5;
"""))
```

### Observation

Technology generated the highest total sales among all product categories. The Canon imageCLASS 2200 Advanced Copier was the top-selling product, while Sean Miller recorded the highest total sales among customers.

```python
print("Data Validation")
display(run_query("""
SELECT
    COUNT(*) AS Total_Rows,
    SUM(CASE WHEN Sales IS NULL THEN 1 ELSE 0 END) AS Null_Sales,
    SUM(CASE WHEN Profit IS NULL THEN 1 ELSE 0 END) AS Null_Profit
FROM superstore;
"""))

print("\nDuplicate Order IDs")
display(run_query("""
SELECT
    `Order ID`,
    COUNT(*) AS Frequency
FROM superstore
GROUP BY `Order ID`
HAVING COUNT(*) > 1
LIMIT 10;
"""))
```

### Observation

The dataset contains 9,694 records with no NULL values in the Sales or Profit columns. Duplicate Order IDs represent multiple products within the same customer order and are expected in transactional sales data.
