```python
print("="*90)
print("SECTION B - FILTERING & OPTIMIZATION")
print("="*90)

# Q7
run_assignment_query(
    "Q7. Orders with status = 'Delivered'",
    """
    SELECT *
    FROM orders
    WHERE status = 'Delivered';
    """
)

# Q8
run_assignment_query(
    "Q8. Electronics products with unit_price > 2000",
    """
    SELECT *
    FROM products
    WHERE category='Electronics'
    AND unit_price>2000;
    """
)

# Q9
run_assignment_query(
    "Q9. Customers from Maharashtra who joined in 2024",
    """
    SELECT *
    FROM customers
    WHERE state='Maharashtra'
    AND YEAR(join_date)=2024;
    """
)

# Q10
run_assignment_query(
    "Q10. Orders placed between 2024-08-10 and 2024-08-25 excluding Cancelled",
    """
    SELECT *
    FROM orders
    WHERE order_date BETWEEN '2024-08-10' AND '2024-08-25'
    AND status<>'Cancelled';
    """
)

print("\n" + "="*90)
print("Q11. Purpose of idx_orders_date")
print("="*90)

print("""
The index 'idx_orders_date' improves the performance of queries
that search, filter or sort records using the order_date column.
It reduces the number of rows scanned and speeds up data retrieval.
""")

print("\n" + "="*90)
print("Q12. SARGable Query")
print("="*90)

print("""
Query:
SELECT * FROM customers
WHERE YEAR(join_date)=2024;

This query is NOT SARGable because YEAR() is applied on the indexed column.

A better version is:

SELECT *
FROM customers
WHERE join_date BETWEEN '2024-01-01' AND '2024-12-31';

This allows MySQL to use an index efficiently.
""")
```

```python
print("="*90)
print("SECTION C - AGGREGATION")
print("="*90)

# Q13
run_assignment_query(
    "Q13. Count the total number of orders",
    """
    SELECT COUNT(*) AS Total_Orders
    FROM orders;
    """
)

# Q14
run_assignment_query(
    "Q14. Total revenue from Delivered orders",
    """
    SELECT SUM(total_amount) AS Total_Revenue
    FROM orders
    WHERE status='Delivered';
    """
)

# Q15
run_assignment_query(
    "Q15. Average unit price by category",
    """
    SELECT
        category,
        ROUND(AVG(unit_price),2) AS Average_Unit_Price
    FROM products
    GROUP BY category;
    """
)

# Q16
run_assignment_query(
    "Q16. Order count and revenue by status",
    """
    SELECT
        status,
        COUNT(*) AS Order_Count,
        ROUND(SUM(total_amount),2) AS Total_Revenue
    FROM orders
    GROUP BY status
    ORDER BY Total_Revenue DESC;
    """
)

# Q17
run_assignment_query(
    "Q17. Most expensive and cheapest product in each category",
    """
    SELECT
        category,
        MAX(unit_price) AS Highest_Price,
        MIN(unit_price) AS Lowest_Price
    FROM products
    GROUP BY category;
    """
)

# Q18
run_assignment_query(
    "Q18. Categories with average unit price greater than ₹2000",
    """
    SELECT
        category,
        ROUND(AVG(unit_price),2) AS Average_Unit_Price
    FROM products
    GROUP BY category
    HAVING AVG(unit_price) > 2000;
    """
)
```

```python
print("="*90)
print("SECTION D - JOINS & RELATIONSHIPS")
print("="*90)

# Q19
run_assignment_query(
    "Q19. Display each order along with customer details",
    """
    SELECT
        o.order_id,
        o.order_date,
        c.first_name,
        c.last_name,
        o.total_amount
    FROM orders o
    INNER JOIN customers c
    ON o.customer_id = c.customer_id;
    """
)

# Q20
run_assignment_query(
    "Q20. Display all customers and their orders (if any)",
    """
    SELECT
        c.customer_id,
        c.first_name,
        c.last_name,
        o.order_id,
        o.order_date,
        o.total_amount
    FROM customers c
    LEFT JOIN orders o
    ON c.customer_id = o.customer_id;
    """
)

# Q21
run_assignment_query(
    "Q21. Display order details with product information",
    """
    SELECT
        o.order_id,
        p.product_name,
        oi.quantity,
        oi.unit_price,
        oi.discount_pct
    FROM orders o
    INNER JOIN order_items oi
    ON o.order_id = oi.order_id
    INNER JOIN products p
    ON oi.product_id = p.product_id;
    """
)

print("\n" + "="*90)
print("Q22. LEFT JOIN vs RIGHT JOIN")
print("="*90)

print("""
LEFT JOIN:
Returns all rows from the left table and matching rows from the right table.

RIGHT JOIN:
Returns all rows from the right table and matching rows from the left table.

FULL OUTER JOIN:
Returns all matching and non-matching rows from both tables.
MySQL does not support FULL OUTER JOIN directly.
It can be simulated using UNION of LEFT JOIN and RIGHT JOIN.
""")

print("\n" + "="*90)
print("Q23. Foreign Key Relationships")
print("="*90)

print("""
Foreign Keys:

orders.customer_id  --> customers.customer_id

order_items.order_id --> orders.order_id

order_items.product_id --> products.product_id

If customer_id = 999 is inserted into the orders table,
MySQL will reject the INSERT statement because of the
FOREIGN KEY constraint.
""")
```

```python
print("="*90)
print("SECTION E - ADVANCED CONCEPTS")
print("="*90)

# Q24
run_assignment_query(
    "Q24. Classify products into price tiers",
    """
    SELECT
        product_name,
        unit_price,
        CASE
            WHEN unit_price < 1000 THEN 'Budget'
            WHEN unit_price BETWEEN 1000 AND 3000 THEN 'Mid-Range'
            ELSE 'Premium'
        END AS price_tier
    FROM products;
    """
)

# Q25
run_assignment_query(
    "Q25. Count Delivered vs Not Delivered orders",
    """
    SELECT
        SUM(CASE WHEN status='Delivered' THEN 1 ELSE 0 END) AS Delivered,
        SUM(CASE WHEN status<>'Delivered' THEN 1 ELSE 0 END) AS Not_Delivered
    FROM orders;
    """
)

print("\n" + "="*90)
print("Q26. ACID Properties")
print("="*90)

print("""
A - Atomicity
A transaction is completed fully or not at all.

C - Consistency
The database remains valid before and after every transaction.

I - Isolation
Transactions execute independently without interfering.

D - Durability
Once committed, changes are permanently stored.

Example:
In a bank transfer, money is deducted from one account
and credited to another. If any step fails, the entire
transaction is rolled back.
""")

print("\n" + "="*90)
print("Q27. SQL Transaction")
print("="*90)

print("""
START TRANSACTION;

INSERT INTO orders
(order_id, customer_id, order_date, status, total_amount)
VALUES
(1011,102,CURDATE(),'Pending',1598.00);

INSERT INTO order_items
(item_id,order_id,product_id,quantity,unit_price,discount_pct)
VALUES
(16,1011,201,1,999.00,0),
(17,1011,203,1,599.00,0);

UPDATE products
SET stock_qty=stock_qty-1
WHERE product_id IN (201,203);

COMMIT;

If any statement fails before COMMIT,
ROLLBACK should be executed.
""")
```
